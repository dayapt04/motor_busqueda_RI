```python
import sys

sys.path.insert(0, "../src")

from similitud import BuscadorJaccard, BuscadorCoseno, mostrar_resultados
import pandas as pd

df = pd.read_csv("../data/processed/corpus_limpio.csv")

jac = BuscadorJaccard()
cos = BuscadorCoseno()
```

```python
consulta = "analista de datos python"

mostrar_resultados(jac.buscar(consulta, top_k=10), "Jaccard", df)
mostrar_resultados(cos.buscar(consulta, top_k=10), "Coseno TF-IDF", df)
```

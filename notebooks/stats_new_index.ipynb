```python
import pandas as pd
import pickle

df = pd.read_csv("../data/processed/corpus_limpio.csv")
print(f"Documentos: {len(df):,}")
print(f"Columnas: {df.columns.tolist()}")

with open("../data/processed/indice_invertido.pkl", "rb") as f:
    datos = pickle.load(f)

indice = datos["indice"]
longitud_docs = datos["longitud_docs"]
total_tokens = sum(longitud_docs.values())
total_entradas = sum(len(v) for v in indice.values())

print(f'Documentos indexados:          {datos["total_documentos"]:,}')
print(f"Total tokens:                  {total_tokens:,}")
print(f"Promedio tokens por doc:       {total_tokens/len(longitud_docs):.1f}")
print(f"Términos únicos:               {len(indice):,}")
print(f"Entradas en el índice:         {total_entradas:,}")
print(f"Promedio docs por término:     {total_entradas/len(indice):.1f}")
```

```python
import sys

sys.path.insert(0, "../src")

from models import (
    cargar_indice,
    BuscadorJaccard,
    BuscadorCoseno,
    BuscadorBM25,
    mostrar_resultados,
)
import pandas as pd

df = pd.read_csv("../data/processed/corpus_limpio.csv")
datos = cargar_indice()
jac = BuscadorJaccard(datos_indice=datos)
cos = BuscadorCoseno(datos_indice=datos)
bm25 = BuscadorBM25(datos_indice=datos)
```

```python
from rich.console import Console
from rich.table import Table
from rich import box

console = Console()


def mostrar_tabla(resultados, modelo, consulta, corpus_df=None):
    tabla = Table(
        title=f'🔍  "{consulta}"',
        box=box.DOUBLE_EDGE,
        show_header=True,
        header_style="bold cyan",
    )
    tabla.add_column("#", style="dim", width=4, justify="right")
    tabla.add_column("Título", width=45)
    tabla.add_column("Score", style="green", width=10, justify="right")
    tabla.caption = f"Modelo: {modelo}  |  {len(resultados)} resultado(s)"

    for i, (doc_id, score) in enumerate(resultados, 1):
        titulo = doc_id[:43] + "..."
        if corpus_df is not None:
            fila = corpus_df[corpus_df["job_id"] == doc_id]
            if not fila.empty:
                titulo = fila.iloc[0]["job_title"]
        tabla.add_row(str(i), titulo, f"{score:.4f}")

    console.print(tabla)
```

```python
consulta = "analista de datos python"

mostrar_tabla(jac.buscar(consulta, top_k=10), "Jaccard", consulta, df)
mostrar_tabla(cos.buscar(consulta, top_k=10), "Coseno TF-IDF", consulta, df)
mostrar_tabla(bm25.buscar(consulta, top_k=10), "BM25", consulta, df)
```
